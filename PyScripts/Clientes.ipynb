```python
import os

# Nome do arquivo do banco de dados
nome_arquivo = "clientes.db"

# Verifica se o arquivo existe
if os.path.exists(nome_arquivo):
    # Exclui o arquivo do banco de dados
    os.remove(nome_arquivo)
    print("Banco de dados excluído com sucesso.")
else:
    print("O banco de dados não existe.")
```

```python
import unqlite

def incluir_clientes():
    # Abrir ou criar o banco de dados UnQLite
    db = unqlite.UnQLite("clientes.db")    

    while True:
        # Obter os dados do novo cliente
        cpf = input("Digite o cpf do cliente (ou 'sair' para encerrar): ")
        if cpf.lower() == "sair":
            break

        # Verificar se o cliente existe no banco de dados
        if str(cpf) in db:
            # Obter o valor associado ao CPF
            cliente_string = db[str(cpf)].decode("utf-8")

            # Converter a string do cliente para dicionário
            cliente = eval(cliente_string)

            # Exibir as informações do cliente
            print("Cliente encontrado:")
            print("CPF:", cliente['cpf'])
            print("Nome:", cliente['nome'])
            print("Email:", cliente['email'])
            break
        else:
            nome = input("Digite o nome do cliente: ")
            email = input("Digite o email do cliente: ")   

        # Criar um dicionário com os dados do cliente
        cliente = {
            "nome": nome,
            "cpf":cpf,
            "email": email
        }

        # Converter o dicionário para string
        cliente_string = str(cliente)

        # Incluir o cliente no banco de dados
        db[str(cpf)] = cliente_string

        print("Cliente cadastrado com sucesso.")

    # Fechar o banco de dados
    db.close()
```

```python
def consultar_cliente():
    # Abrir o banco de dados UnQLite
    db = unqlite.UnQLite("clientes.db")
   
    # Solicitar o CPF do cliente a ser consultado
    cpf_consulta = input("Digite o CPF do cliente a ser consultado: ")
    
    # Verificar se o cliente existe no banco de dados
    if str(cpf_consulta) in db:
        # Obter o valor associado ao CPF
        cliente_string = db[str(cpf_consulta)].decode("utf-8")

        # Converter a string do cliente para dicionário
        cliente = eval(cliente_string)

        # Exibir as informações do cliente
        print("Cliente encontrado:")
        print("Nome:", cliente['nome'])
        print("CPF:", cliente['cpf'])
        print("Email:", cliente['email'])
    else:
        print("Cliente não encontrado.")

    # Fechar o banco de dados
    db.close()
```

```python
def exibir_clientes():
    # Abrir o banco de dados UnQLite
    db = unqlite.UnQLite("clientes.db")

    # Percorrer todos os registros do banco de dados
    for chave, valor in db.items():
        # Converter a string do cliente para dicionário
        cliente = eval(valor.decode("utf-8"))

        # Exibir as informações do cliente
        print("Cliente:")
        print("Chave:", chave)
        print("Nome:", cliente['nome'])
        print("CPF:", cliente['cpf'])
        print("Email:", cliente['email'])
        print("------------------")

    # Fechar o banco de dados
    db.close()
```

```python
import unqlite

def listar_registros():
    # Abrir o banco de dados UnQLite
    db = unqlite.UnQLite("clientes.db")

    # Obter todas as chaves (IDs) do banco de dados
    chaves = db.keys()

    # Percorrer todas as chaves e exibir os registros correspondentes
    for chave in chaves:
        registro = db[str(chave)]
        print(f"ID: {chave}")
        print(f"Registro: {registro}")
        print("-" * 20)

    # Fechar o banco de dados
    db.close()

def main():
    listar_registros()

if __name__ == "__main__":
    main()
```

```python
# Menu principal
while True:
    print("1 - Incluir clientes")
    print("2 - Consultar cliente")
    print("3 - Exibir clientes")
    print("4 - Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        incluir_clientes()
    elif opcao == "2":
        consultar_cliente()
    elif opcao == "3":
        exibir_clientes()
    elif opcao == "4":
        break
    else:
        print("Opção inválida. Tente novamente.\n")
```

```python
import unqlite
import random
import string

def gerar_cpf():
    # Gera um CPF aleatório com números de 0 a 9
    cpf = ''.join(random.choices(string.digits, k=11))
    return cpf

def gerar_nome():
    # Gera um nome aleatório com letras maiúsculas e minúsculas
    nome = ''.join(random.choices(string.ascii_letters, k=random.randint(5, 10)))
    return nome

def gerar_email():
    # Gera um email aleatório com letras minúsculas
    email = ''.join(random.choices(string.ascii_lowercase, k=random.randint(5, 10))) + "@example.com"
    return email

def incluir_registros():
    # Abrir o banco de dados UnQLite
    db = unqlite.UnQLite("clientes.db")

    # Incluir 500 registros de clientes
    for _ in range(500):
        cpf = gerar_cpf()
        nome = gerar_nome()
        email = gerar_email()

        # Criar um dicionário com os atributos do cliente
        cliente = {
            'cpf': cpf,
            'nome': nome,
            'email': email
        }

        # Converter o dicionário para string
        cliente_string = str(cliente)

        # Incluir o cliente no banco de dados
        db[str(cpf)] = cliente_string

    # Fechar o banco de dados
    db.close()

def main():
    # Incluir os registros de clientes no banco de dados
    incluir_registros()

    print("Registros incluídos com sucesso!")

if __name__ == "__main__":
    main()
```
